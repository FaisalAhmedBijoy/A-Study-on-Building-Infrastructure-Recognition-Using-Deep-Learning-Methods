```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Train accuracy.csv')
df.head()
```

```python
plt.plot(df['epochs'],df['Sequential CNN Acc'],label='Seq. Acc',marker= '.',markersize=6)
plt.plot(df['epochs'],df['Sequential CNN Val Acc'],label='Seq. Val Acc',marker='v',markersize=6)
plt.plot(df['epochs'],df['Hybrid CNN Acc'],label='Hyb. Acc',marker='^',markersize=6)
plt.plot(df['epochs'],df['Hybrid CNN Val Acc'],label='Hyb. Val Acc',marker='x',markersize=6)
plt.legend()
#plt.title('Train Accuracy Graph')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.savefig('EictTrain.png')
```

```python
plt.plot(df['epochs'],df['Sequential CNN Acc'],label='Sequential CNN Train Accuracy')
plt.plot(df['epochs'],df['Sequential CNN Val Acc'],label='Sequential CNN Validation Accuracy')
plt.plot(df['epochs'],df['Hybrid CNN Acc'],label='Hybrid CNN Train Accuracy')
plt.plot(df['epochs'],df['Hybrid CNN Val Acc'],label='Hybrid CNN Validation Accuracy')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.savefig('Train Graph Without Title.png')
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['epochs'],df['Sequential CNN Acc'],label='Sequential CNN Train Accuracy')
plt.plot(df['epochs'],df['Sequential CNN Val Acc'],label='Sequential CNN Validation Accuracy')
plt.plot(df['epochs'],df['Hybrid CNN Acc'],label='Hybrid CNN Train Accuracy')
plt.plot(df['epochs'],df['Hybrid CNN Val Acc'],label='Hybrid CNN Validation Accuracy')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.savefig('Train Graph Without Title Fig size 8x5.png')
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['epochs'],df['Sequential CNN Acc'],label='Sequential CNN Acc')
plt.scatter(df['epochs'],df['Sequential CNN Val Acc'],label='Sequential CNN Val Acc')
plt.scatter(df['epochs'],df['Hybrid CNN Acc'],label='Hybrid CNN Acc')
plt.scatter(df['epochs'],df['Hybrid CNN Val Acc'],label='Hybrid CNN Val Acc')
plt.legend()
plt.title('Train Accuracy Graph')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.savefig('Train Graph Scatter.png')
```
