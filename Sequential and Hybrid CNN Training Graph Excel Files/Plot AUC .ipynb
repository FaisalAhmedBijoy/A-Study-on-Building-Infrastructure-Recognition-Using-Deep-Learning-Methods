```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('AUC train.csv')
df.head()
```

```python
plt.plot(df['epochs'],df['Sequential CNN AUC'],label='Seq. AUC',marker= '.',markersize=6)
plt.plot(df['epochs'],df['Sequential CNN Val AUC'],label='Seq. Val AUC',marker='v',markersize=6)
plt.plot(df['epochs'],df['Hybrid CNN AUC'],label='Hyb. AUC',marker='^',markersize=6)
plt.plot(df['epochs'],df['Hybrid CNN Val AUC'],label='Hyb. Val AUC',marker='x',markersize=6)
plt.legend()
#plt.title('Train Accuracy Graph')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.savefig('EictAUC.png')
```

```python
plt.plot(df['epochs'],df['Sequential CNN AUC'],label='Sequential CNN AUC')
plt.plot(df['epochs'],df['Sequential CNN Val AUC'],label='Sequential CNN Val AUC')
plt.plot(df['epochs'],df['Hybrid CNN AUC'],label='Hybrid CNN AUC')
plt.plot(df['epochs'],df['Hybrid CNN Val AUC'],label='Hybrid CNN Val AUC')
plt.legend()
plt.title('Train Accuracy Graph')
plt.xlabel('Epochs')
plt.ylabel('AUC')
plt.savefig('AUC Train Graph without title.png')
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['epochs'],df['Sequential CNN AUC'],label='Sequential CNN AUC')
plt.plot(df['epochs'],df['Sequential CNN Val AUC'],label='Sequential CNN Val AUC')
plt.plot(df['epochs'],df['Hybrid CNN AUC'],label='Hybrid CNN AUC')
plt.plot(df['epochs'],df['Hybrid CNN Val AUC'],label='Hybrid CNN Val AUC')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('AUC')
plt.savefig('AUC Train Graph without title 8x5.png')
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['epochs'],df['Sequential CNN AUC'],label='Sequential CNN AUC')
plt.scatter(df['epochs'],df['Sequential CNN Val AUC'],label='Sequential CNN Val AUC')
plt.scatter(df['epochs'],df['Hybrid CNN AUC'],label='Hybrid CNN AUC')
plt.scatter(df['epochs'],df['Hybrid CNN Val AUC'],label='Hybrid CNN Val AUC')
plt.legend()
plt.title('Train Accuracy Graph')
plt.xlabel('Epochs')
plt.ylabel('AUC')
plt.savefig('AUC Train Graph without title Scatter Plot 8x5.png')
```
