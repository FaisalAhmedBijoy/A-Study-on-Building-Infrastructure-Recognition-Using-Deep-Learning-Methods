## Precision Recall Calculation

```python
new_df=df
```

```python
new_df.sort_values(by='Score',ascending=False)
new_df.head()
```

```python
threshold_score=0.8
```

```python
new_df.loc[new_df['Score']>=threshold_score,'TP']=1.0
new_df.loc[new_df['Score']<threshold_score,'FP']=1.0
```

```python
new_df.tail(30)
```

```python
new_df=new_df.fillna(0)
```

```python
new_df.head()
```

```python
new_df['Acc TP']=new_df.TP.cumsum()
```

```python
new_df.head()
```

```python
new_df['Acc FP']=new_df.FP.cumsum()
```

```python
new_df.head()
```

```python
new_df['precision']=new_df['Acc TP']/(new_df['Acc TP']+new_df['Acc FP'])
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['recall']=new_df['Acc TP']/185
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df.describe()
```

```python
plt.scatter(new_df['precision'],new_df['recall'])
plt.xlabel('precision')
plt.ylabel('recall')
```

```python
plt.scatter(new_df['recall'],new_df['precision'])
plt.xlabel('recall')
plt.ylabel('precision')
plt.title('Precision VS Recall')
```
