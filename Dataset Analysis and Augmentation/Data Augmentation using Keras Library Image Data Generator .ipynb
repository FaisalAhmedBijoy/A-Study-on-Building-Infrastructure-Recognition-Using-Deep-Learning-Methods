# Data Augmentation
- horizontal and vertical shift
- horizontal and vertical flip
- zoom range
- rotation range
- height and width shift
- brightness change

## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import PIL
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
```

## Load Image

```python
img=load_img('mosque35.png')
plt.imshow(img)
```

```python
type(img)
```

```python
# img to array
image_array=np.array(img)
image_array.shape
```

## Convert 4D Image

```python
test_image=np.array([image_array])
test_image.shape
```

# Horizontal shift

```python
horizontal_shift_datagen=ImageDataGenerator(height_shift_range=0.5)
it=horizontal_shift_datagen.flow(test_image,batch_size=1)
print(it)
print('Iterator shpae: ',it.next().shape)
```

```python
fig=plt.figure(figsize=(15,15))
rows=3
cols=3
for i in range(rows*cols):
    fig.add_subplot(rows,cols,i+1)
    image=it.next()
    plt.imshow(image[0].astype('uint8'))
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```

## Vertical Shift

```python
vertical_shift_datagen=ImageDataGenerator(width_shift_range=0.5)
it=vertical_shift_datagen.flow(test_image,batch_size=1)
print(it)
print('Iterator shpae: ',it.next().shape)
```

```python
fig=plt.figure(figsize=(15,15))
rows=3
cols=3
for i in range(rows*cols):
    fig.add_subplot(rows,cols,i+1)
    image=it.next()
    plt.imshow(image[0].astype('uint8'))
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```

## Rotation 

```python
rotation_datagen=ImageDataGenerator(rotation_range=45)
it=rotation_datagen.flow(test_image,batch_size=1)
print(it)
print('Iterator shpae: ',it.next().shape)

fig=plt.figure(figsize=(10,10))
rows=2
cols=2
for i in range(rows*cols):
    fig.add_subplot(rows,cols,i+1)
    image=it.next()
    plt.imshow(image[0].astype('uint8'))
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```

## Brightness 

```python
brightness_datagen=ImageDataGenerator(brightness_range=[0.2,1.0])
it=brightness_datagen.flow(test_image,batch_size=1)
print(it)
print('Iterator shpae: ',it.next().shape)

fig=plt.figure(figsize=(10,10))
rows=2
cols=2
for i in range(rows*cols):
    fig.add_subplot(rows,cols,i+1)
    image=it.next()
    plt.imshow(image[0].astype('uint8'))
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```

 ## Apply all data augmentation

```python
datagen = ImageDataGenerator(
          rotation_range=40,
          width_shift_range=0.2,
          height_shift_range=0.2,
          shear_range=0.2,
          zoom_range=0.2,
          horizontal_flip=True,
          fill_mode='nearest')

it=datagen.flow(test_image,batch_size=1)
print(it)
print('Iterator shpae: ',it.next().shape)

fig=plt.figure(figsize=(10,10))
rows=3
cols=3
for i in range(rows*cols):
    fig.add_subplot(rows,cols,i+1)
    image=it.next()
    plt.imshow(image[0].astype('uint8'))
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```
