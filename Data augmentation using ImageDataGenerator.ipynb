```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing import image
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=image.load_img('mosque35.png')
plt.imshow(img)
```

```python
img
```

```python
type(img)
```

```python
datagen=image.ImageDataGenerator()
datagen
```

```python
#convert numpy array
data=image.img_to_array(img)
# expand the dimension
samples=np.expand_dims(data,0)
```

## Horizontal and vertical shift

```python
shift_datagen=image.ImageDataGenerator(width_shift_range=[-100,100])
it=shift_datagen.flow(samples,batch_size=1)
print(it)

for i in range(2):
    plt.subplot(120+1+i)
    
    batch=it.next()
    img=batch[0].astype('uint8')
    plt.imshow(img)
plt.show()
```

```python
shift_datagen=image.ImageDataGenerator(height_shift_range=0.5)
it=shift_datagen.flow(samples,batch_size=1)
print(it)
for i in range(2):
    plt.subplot(120+1+i)
    batch=it.next()
    img=batch[0].astype('uint8')
    plt.imshow(img)
    figure_name='shift'+str(i)+'.png'
    plt.savefig(figure_name)
plt.show()
```

## Horizontal and vertical flip augmentation

```python
flip_datagen=image.ImageDataGenerator(horizontal_flip=True)
it=flip_datagen.flow(samples,batch_size=1)
print(it)
for i in range(2):
    plt.subplot(120+1+i)
    batch=it.next()
    img=batch[0].astype('uint8')
    plt.imshow(img)
    figure_name='flip'+str(i)+'.png'
    plt.savefig(figure_name)
plt.show()
```

```python
flip_datagen=image.ImageDataGenerator(vertical_flip=True)
it=flip_datagen.flow(samples,batch_size=1)
print(it)
for i in range(2):
    plt.subplot(120+1+i)
    batch=it.next()
    img=batch[0].astype('uint8')
    plt.imshow(img)
    figure_name='vertical flip'+str(i)+'.png'
    plt.savefig(figure_name)
plt.show()
```

## Random rotation augmentation

```python
rotation_datagen=image.ImageDataGenerator(rotation_range=90)
it=rotation_datagen.flow(samples,batch_size=1)
print(it)
for i in range(2):
    plt.subplot(120+1+i)
    batch=it.next()
    img=batch[0].astype('uint8')
    plt.imshow(img)
    figure_name='rotation change'+str(i)+'.png'
    plt.savefig(figure_name)
plt.show()
```

## Random brightness augmentation

```python
brightness_datagen=image.ImageDataGenerator(brightness_range=[0.2,1.0])
it=brightness_datagen.flow(samples,batch_size=1)
print(it)
for i in range(2):
    plt.subplot(120+1+i)
    plt.figsize=(20,20)
    batch=it.next()
    img=batch[0].astype('uint8')
    plt.imshow(img)
    figure_name='brightness'+str(i)+'.png'
    plt.savefig(figure_name)
plt.show()
```

## Random zoom augmentation

```python
zoom_datagen=image.ImageDataGenerator(zoom_range=[0.5,1.0])
it=zoom_datagen.flow(samples,batch_size=1)
print(it)
for i in range(2):
    plt.subplot(120+1+i)
    batch=it.next()
    img=batch[0].astype('uint8')
    plt.imshow(img)
    figure_name='zoom range'+str(i)+'.png'
    plt.savefig(figure_name)
plt.show()
```
